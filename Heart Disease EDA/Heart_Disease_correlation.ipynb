### **Correlation Heatmap**

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/heart.csv')
```

```python
df
```

```python
data = df.select_dtypes(include = np.number)
```

```python
data
```

```python
data.corr()
```

```python
sns.heatmap(data.corr())
```

```python
sns.heatmap(data.corr(), annot = True)
```

### **Correlation Jointplot**

```python
# Jointplot give me both distribution & scatterplot
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df)
```

```python
sns.jointplot(x = 'Age', y = 'Age', data = df)
```

```python
# correlation between heartdisese and maxhr
sns.jointplot(x = 'HeartDisease', y = 'MaxHR', data = df)
```

```python
# correlation between heartdisese and oldpeak
sns.jointplot(x = 'HeartDisease', y = 'Oldpeak', data = df)
```

```python
# correlation between cholesterol and oldpeak
sns.jointplot(x = 'Cholesterol', y = 'Oldpeak', data = df)
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind = 'reg')
```

```python
sns.jointplot(x = 'Cholesterol', y = 'FastingBS', data = df, kind = 'reg')
```
