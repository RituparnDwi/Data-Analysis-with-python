##**Finding Outliers Using IQR**

```python
Q1 = num_col.quantile(0.25)
```

```python
print(Q1)
```

```python
Q3 = num_col.quantile(0.75)
```

```python
print(Q3)
```

```python
IQR = Q3 - Q1
```

```python
print(IQR)
```

```python
# The values which are below Q1 - 1.5*IQR and above Q3 + 1.5*IQR are Outliers.
```

```python
min = Q1 - 1.5*IQR
max = Q3 + 1.5*IQR
```

```python
print(min)
```

```python
print(max)
```

```python
Outliers = num_col[(num_col < min) | (num_col > max)]
```

```python
Outliers
```

```python
Outliers['Delinquent'].value_counts()
```

```python
Outliers.count()
```

```python
min_lims = num_col.quantile(0.01)
max_lims = num_col.quantile(0.99)
```

```python
print(min_lims)
```

```python
print(max_lims)
```

```python
Outliers_ = num_col[(num_col < min_lims) | (num_col > max_lims)]
```

```python
Outliers_.count()
```

```python
Trimmed_num_col = num_col[(num_col < max_lims) & (num_col > min_lims)]
```

##**Finding Outlier Using Z-score**

```python
mean = num_col.mean()
```

```python
std = num_col.std()
```

```python
print(mean)
```

```python
print(std)
```

```python
High_value = mean + 3 * std
```

```python
Low_value = mean - 3 * std
```

```python
print(High_value)
```

```python
print(Low_value)
```

```python
outlier = num_col[(num_col > High_value) | (num_col < Low_value)]
```
