```python
import pandas as pd
```

#**Creating a DataFrame Using List**#

```python
lst = ['Geeks', 'for', 'Geeks', 'is', 'portal', 'for', 'Geeks']
pd.DataFrame(lst)
```

```python
lst = [['Geeks', '25'], ['for', '30'], ['Geeks', '20']]
pd.DataFrame(lst, columns = ['Tag', 'number'])
```

###**Creating DataFrame using Dictionary**###

```python
data = {'Name':['Tom', 'nick', 'krish', 'jack'], 'Age':[20, 21, 19, 18]}
pd.DataFrame(data)
```

##**Column Selection**##

##**Slicing in DataFrame Using iloc and loc**

```python
data3 = {'One' : pd.Series([1,2,3,4,5,6]),
         'Two' : pd.Series([10,20,30,40,50,60]),
         'Three' : pd.Series([100,200,300,400,500,600]),
         'Four' : pd.Series([1000,2000,3000,4000,5000,6000])}
df1 = pd.DataFrame(data3)
df1
```

loc Operation

```python
df1.loc[2]
```

```python
df1.loc[2:]
```

```python
df1.loc[1:2]
```

```python
df1.loc[2:3, ['One', 'Three']]
```

```python
df1.loc[2:3, 'One':'Three']
```

```python
df1.loc[:, ['One', 'Three']]
```

```python
df1.loc[1:3, :]
```

```python
df1.loc[1: , 'One':'Three']
```

```python
df1.loc[1: , 'One': ]
```

```python
df1.loc[1:]
```

```python
df1.loc[1:-1]
```

```python
df1.loc[1:5:2]
```

iloc Operation

```python
df1.iloc[1:2]
```

```python
df1.iloc[1:3]
```

```python
df1.iloc[1: , 1: ]
```

```python
df1.iloc[ : , 2: ]
```

```python
df1.iloc[1:-1, 1:-1]
```

```python
df1.iloc[0,2] # Selecting specific data
```

```python
df1.iloc[[0,2]] # Selecting only that row which i have passed by using comma
```
