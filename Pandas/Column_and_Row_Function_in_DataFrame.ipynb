#**Column Addition in DataFrame**

We can add column in DataFrame in many ways....
1. Using List
2. Using Pandas Series
3. Using an existing column

```python
import pandas as pd
```

```python
data = {'One'  : [1,2,3,4,5,6],
       'Two'   : [10,20,30,40,50,60],
       'Three' : [100,200,300,400,500,600],
       'Four'  : [1000,2000,3000,4000,5000,6000]}
df = pd.DataFrame(data)
df
```

**Using List**

```python
l = [11,22,33,44,55,66]
df['Five'] = l
df
```

**Using Pandas Series**

```python
sr = pd.Series([111,222,333,444,555,666])
df['Six'] = sr
df
```

**Using Existing Column**

```python
df['Seven'] = df['Four'] + 10
```

```python
df
```

#**Column Deletion in DataFrame**

We can Delete column using....
1. Using del f/c
2. Using drop f/c

**Using del f/c**

```python
del df['Seven']
df
```

**Using drop f/c**

```python
df.drop('Six',axis=1,inplace=True)
df
```

#**Addition of Row**

We can add row by concat method

```python
df1 = pd.DataFrame([[11,22,33,44,55,66]],columns=['One','Two','Three','Four','Five','Six'])
df2 = pd.DataFrame([[111,222,333,444,555,666]],columns=['One','Two','Three','Four','Five','Six'])
df3 = pd.concat([df1,df2])
```

```python
df3
```

```python
df3 = df3.reset_index(drop=True)
```

```python
df3
```
