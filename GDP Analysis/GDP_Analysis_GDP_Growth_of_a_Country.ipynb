##**Analysing the Arab World**

```python
df[df['Country Name'] == 'Arab World']
```

```python
df_pr = df[df['Country Name'] == 'Arab World']
```

```python
df_pr.plot(kind='line', x='Year', y='Value',
           figsize=(12, 4),
           grid='True',
           title='Arab World GDP',
           xlabel='Years',
           ylabel='GDP')
```

```python
data = df_pr.values

for i in range(len(data)):
  print(data[i])
```

```python
GDP_Change = [0]

for i in range(1, len(data)):

  prev_year = data[i-1][3]
  current_year = data[i][3]

  GDP_Change.append(round((current_year - prev_year) / prev_year * 100,2))
```

```python
GDP_Change
```

```python
df_pr.assign(GDP=GDP_Change)
```

##**Analysing GDP of India**

```python
df_pr1 = df[df['Country Name'] == 'India']
```

```python
df_pr1.plot(kind='line', x='Year', y='Value',
            figsize=(12,6),
            xlabel='Years',
            ylabel='GDP',
            grid='True',
            title='India GDP'
            )
```

```python
gdp_change = [0]

data1 = df_pr1.values

for i in range(1, len(data1)):

  prev_year = data1[i-1][3]
  current_year = data1[i][3]

  gdp_change.append(round((current_year - prev_year) / prev_year * 100,2))
```

```python
gdp_change
```

```python
df_pr1.assign(India_GDP=gdp_change)
```
