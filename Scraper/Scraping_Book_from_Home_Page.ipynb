```python
data = []

for sp in soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3'):

  # Scrap the Book Title
  Title = sp.find_all('a')[-1].get('title')

  # Scrap the Book link
  Book_link = sp.find_all('a')[-1].get('href')

  # Scrap the image link
  Img_link = sp.find('img').get('src')

  # Scrap the Book rating
  Rating = sp.find('p').get('class')[-1]

  # Scrap the Book price
  Price = sp.find('p', class_='price_color').text[1:]

  # Scrap the stock
  Instock_availability = sp.find('p', class_='instock availability').text.strip()

  data.append([Title, Rating, Price, Instock_availability, Book_link, Img_link])
```

```python
data = []

for sp in soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3'):

  # Scrap the Book Title
  Title = sp.find_all('a')[-1].get('title')

  # Scrap the Book link
  Book_link = 'https://books.toscrape.com/' + sp.find_all('a')[-1].get('href')

  # Scrap the image link
  Img_link = 'https://books.toscrape.com/' + sp.find('img').get('src')

  # Scrap the Book rating
  Rating = sp.find('p').get('class')[-1]

  # Scrap the Book price
  Price = sp.find('p', class_='price_color').text[1:]

  # Scrap the stock
  Instock_availability = sp.find('p', class_='instock availability').text.strip()

  data.append([Title, Rating, Price, Instock_availability, Book_link, Img_link])
```
