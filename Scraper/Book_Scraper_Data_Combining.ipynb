```python
import pandas as pd
```

```python
df_1 = pd.read_csv('/content/books.csv')
df_2 = pd.read_csv('/content/books_data.csv')
```

### **Adding Column in Empty DataFrame**

```python
df = pd.DataFrame()
```

```python
df['Title']             = df_1['Title']
df['UPC']               = df_2['UPC']
df['Type of Book']      = df_2['Type of Book']
df['Price Without Tax'] = df_2['Price Without Tax']
df['Price With Tax']    = df_2['Price With Tax']
df['Tax']               = df_2['Tax']
df['Ratings']           = df_1['Rating']
df['Reviews']           = df_2['Number of Review']
df['Stock']             = df_1['Stock']
df['Availability']      = df_2['Availability']
df['Book Link']         = df_1['Book Link']
df['Image Link']        = df_1['Image Link']
```

```python
df.sample(5)
```

```python
df.to_csv('Final_book_data.csv', index = False)
```
