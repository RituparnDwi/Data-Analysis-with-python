# Scraping Quotes with Multiple Pages

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://quotes.toscrape.com/')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
for page in range(2, 11):
    print(page)
```

```python
for page in range(2, 11):
    print('https://quotes.toscrape.com/page/' + str(page))
```

```python
# Get the progress of your work
from tqdm import tqdm
```

```python
New_data = []

# How much your work have done
for page in tqdm(range(1, 11)):
    
    link = 'https://quotes.toscrape.com/page/' + str(page)
    ress = requests.get(link)
    soup = BeautifulSoup(ress.text, 'html.parser')

    # Go to every quote at every page
    for sp in soup.find_all('div', class_ = 'quote'): 
        quote     = sp.find('span', class_ = 'text')
        author    = sp.find('small').text
        author_id = sp.find('a').get('href')

        tags      = []
        for tag in sp.find_all('a', class_ = 'tag'):
            tags.append(tag.text)
        tags      = ','.join(tags)

        New_data.append([quote, author, author_id, tags])

New_data
```

```python
len(New_data)
```

### Create Data Frame

```python
df = pd.DataFrame(New_data, columns = ['quote', 'author', 'author_id', 'tags'])
```

```python
df['author_link'] = 'https://quotes.toscrape.com' + df['author_id']
```
